## Part 2

Please recreate the bar chart seen below.  

* The figure should be 8x6
* The two colors used are `#74b9ff` and `#7ed6df`


```python
# Use this data to generate the plot!
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]

mplt.style.use("fivethirtyeight")
mplt.figure(figsize=(8, 6))
mplt.bar(resorts, snow_next_24, bottom=snow_last_24, color="#7ed6df", label="Next 24hrs")
mplt.bar(resorts, snow_last_24, color="#74b9ff", label="Last 24hrs")
mplt.title("CO Snow Storm Report")
mplt.ylabel("Inches")
mplt.legend()
mplt.show()
```

## Part 3

Recreate the pie chart seen below:

* I've provided lists for the actual numbers, the labels, and the colors
* Figure should be 8x8
* Notice the percentages in each slice!

```python
# Use these lists to generate the chart!
labels = ["Building Construction", "Deferred Maintenance", "General Contractor Fees", "Owner Costs", "Site Construction"]
costs = [47945975, 3753584, 7340612, 18247137, 5766685]
colors = ["#355070", "#6d597a", "#b56576", "#e56b6f", "#eaac8b"]

mplt.style.use("fivethirtyeight")
mplt.figure(figsize=(8, 8))
mplt.pie(costs, labels=labels, colors=colors, autopct="%1.0f%%", explode=(0, 0.1, 0, 0, 0.1))
mplt.title("School District Cost Breakdown")
mplt.show()
```

## Part 4

Recreate the figure seen in this image:

* Use the `kc_house_data.csv` dataset (import it with Pandas)
* The figure is 10x4

```python
# Do your work here!
import pandas as pd
import numpy as np

df = pd.read_csv("kc_house_data.csv", parse_dates=["date"])
sales_by_week_day = df["date"].dt.dayofweek.value_counts().sort_index()
sales_by_month = df["date"].dt.month.value_counts().sort_index()
mplt.style.use("fivethirtyeight")
mplt.figure(figsize=(10, 4))

mplt.subplot(1, 2, 1)
mplt.title("Sales By Week Day")
mplt.bar(sales_by_week_day.index, sales_by_week_day.values)
mplt.xticks(ticks=np.arange(7), labels=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])

mplt.subplot(1, 2, 2)
mplt.title("Sales By Month")
mplt.ylabel("Home Sales")
mplt.xlabel("Month")
mplt.plot(sales_by_month.index, sales_by_month.values)
mplt.xticks(ticks=np.arange(1, 13))

mplt.show()
```
