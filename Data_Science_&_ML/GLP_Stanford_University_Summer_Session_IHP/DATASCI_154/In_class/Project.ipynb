```python
covid_phase1_tweets_df = pd.read_csv(f"drive/My Drive/{FOLDERNAME}/covid19_apr-jun2020_tweets.csv")
covid_phase2_tweets_df = pd.read_csv(f"drive/My Drive/{FOLDERNAME}/covid19_aug-sep2020_tweets.csv")
covid_phase3_tweets_df = pd.read_csv(f"drive/My Drive/{FOLDERNAME}/covid19_apr-jun2021_tweets.csv")
```

```python
def get_platform(text):
    if 'Android' in text:
        return 'android'
    elif 'iPhone' in text:
        return 'iphone'
    elif 'Web App' in text:
        return 'web app'
    elif 'iPad' in text:
        return 'ipad'
    else:
        # Extract the value between anchor tags
        match = re.search(r'>(.*?)<', text)
        if match:
            return match.group(1).lower()
        else:
            return text.lower()

for phase_df in [covid_phase1_tweets_df, covid_phase2_tweets_df, covid_phase3_tweets_df]:
  phase_df.dropna(subset=["lang"], how="all", inplace=True)
  phase_df.drop(columns=["id", "lang", "compound", "neg", "neu", "pos", "sentiment"], inplace=True)
  phase_df['source'] = phase_df['source'].astype(str).apply(get_platform)
  phase_df.rename(mapper={"source": "platform"}, axis=1, inplace=True)

covid_phase1_tweets_df
```
